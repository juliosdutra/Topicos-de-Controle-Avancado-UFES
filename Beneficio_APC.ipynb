Simulando o benefício de uma estratégia de Controle Avançado

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
valor_medio = 50
delta = 20
limite = 75
N_pontos = 101
controle_classico = valor_medio + 3*np.random.randn(N_pontos)
controle_avancado = valor_medio + np.random.randn(N_pontos)
controle_avancado_otim = (valor_medio + delta) + np.random.randn(N_pontos)
```

```python
plt.figure(figsize=(10,3))
tempo = range(3*N_pontos)
resultado = np.hstack([controle_classico, controle_avancado, controle_avancado_otim])
resultado_medio = np.hstack([valor_medio*np.ones(N_pontos), valor_medio*np.ones(N_pontos),
                             (valor_medio+delta)*np.ones(N_pontos)])
plt.plot(tempo,resultado_medio, 'g', label = 'Valor médio')
plt.plot(tempo,resultado, 'b', label = 'Desempenho')
plt.axhline(limite, color = 'red', label = 'Limite operacional', linestyle = '--')
plt.grid()
plt.legend(loc='upper left')
plt.ylim([30,80])
```
